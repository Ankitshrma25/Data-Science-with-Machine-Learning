## Diabetes Predictions

```python
from sklearn.datasets import load_diabetes
```

```python
dataset=load_diabetes()
```

```python
dataset
```

```python
print(dataset['DESCR'])
```

```python
## Creating Dataframe
import pandas as pd
df_diabetes=pd.DataFrame(dataset.data,columns=['age',
  'sex',
  'bmi',
  'bp',
  's1',
  's2',
  's3',
  's4',
  's5',
  's6'])
df_diabetes.head()
```

```python
## slpliting dataset based on independent and dependent variables
X=df_diabetes
y=dataset['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
## correlation

X_train.corr()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## visualize correlation using heatmap
import seaborn as sns
plt.Figure(figsize=(30,20))
sns.heatmap(X_train.corr(),annot=True)
```

```python
## Applying decision tree regressor
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor.fit(X_train,y_train)
```

## Hyperparamter tunning

```python
param = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter': ['best', 'random'],
    'max_depth': [1,2,3,4,5,10,15,20,25],
    'max_features':['auto', 'sqrt', 'log2']
}
```

```python
## Hyperparameter tunning usig gridsearchCV
from sklearn.model_selection import GridSearchCV
regressor=DecisionTreeRegressor()
grid=GridSearchCV(regressor,param_grid=param,cv=5,scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error,confusion_matrix
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
## Applying decision tree regressor after getting best params
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='squared_error',
 max_depth= 2,
 max_features='log2',
 splitter='best')
regressor.fit(X_train,y_train)
```

```python
## visualization of the decision tree classifier
from sklearn import tree
plt.figure(figsize=(10,15))
tree.plot_tree(regressor,filled=True)
```
