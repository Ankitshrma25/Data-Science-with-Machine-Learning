### Data Visualization Using Seaborn

Seaborn is a library for making statistical graphics in Python. It builds on top of matplotlib and integrates closely with pandas data structures.

```python
import seaborn as sns
```

```python
### Basic ploting with seaborn
tips=sns.load_dataset('tips')
tips
```

```python
## creating a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips,color='red')
plt.title("Scatterplot")
```

```python
### Line plot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line plot for restaurant")
```

```python
sns.set_theme()

sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
dots=sns.load_dataset('dots')
```

```python
import pandas as pd
dots.head()
```

```python
sns.relplot(
    data=dots, kind='line',
    x='time', y='firing_rate', col='align',
    hue='choice', size='coherence', style='choice',
    facet_kws=dict(sharex=False),
)
```

```python
## CAtegorical Plots
## Bar Plot
sns.barplot(x='day', y='total_bill', data=tips)
plt.title("Bar plot of Total Bill By Day")
```

```python
## Distribution Representation
sns.displot(data=tips, x='total_bill', col='time', kde=True)
```

```python
sns.displot(data=dots, x='firing_rate', col='coherence', kde=True)
```

```python
## Box Plot
sns.boxenplot(x='day', y='total_bill', data=tips)
```

```python
## Pair Plot

penguins=sns.load_dataset('penguins')
sns.pairplot(penguins)
```

```python
sns.pairplot(tips)
```

```python
## Heatmap
corr=tips[['total_bill','tip', 'size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
